```python
def input_data():
    data = input("Enter numbers separated by space: ")
    list_data = [int(x) for x in data.split()]
    return list_data

def data_summary(data):
    print("\nData Summary:")
    print(f"- Total elements: {len(data)}")
    print(f"- Minimum value: {min(data)}")
    print(f"- Maximum value: {max(data)}")
    print(f"- Sum of all values: {sum(data)}")
    print(f"- Average value: {round(sum(data)/len(data), 2)}")

def factorial(n):
    if n == 1:
        return 1  
    return n* factorial(n-1)

def calculate_factorial():
    num = int(input("Enter a number for factorial:"))
    print(f"Factorial of {num} is {factorial(num)}")

def filter_data(data):
    value = int(input("Find greater number than this number: "))
    filtered = list(filter(lambda y: y > value, data))
    print(f"Data values greater than {value}: {filtered}")

def sort_data(data):
    print("1. Ascending")
    print("2. Descending")
    option = input("select option: ")
    if option == '1':
        print("Sorted:", sorted(data))
    elif option == '2':
        print("Sorted:", sorted(data, reverse=True))
    else:
        print("Invalid choice")

def dataset_statistics(data):
    if not data:
        print("Dataset is empty. Cannot compute statistics.")
        return

    total_elements = len(data)
    minimum = min(data)
    maximum = max(data)
    total_sum = sum(data)
    average = round(sum(data) / len(data), 2)

    print("Dataset Statistics:")
    print(f"- Total elements: {total_elements}")
    print(f"- Minimum value: {minimum}")
    print(f"- Maximum value: {maximum}")
    print(f"- Sum of all values: {total_sum}")
    print(f"- Average value: {average}")

print("Welcome to the Data Analyzer and Transformer Program!")

def menuu():
    data = []
    while True:

        print("Main Menu :")
        print("1. Input Data")
        print("2. Display Data Summary")
        print("3. Calculate Factorial")
        print("4. Filter Data Threshold")
        print("5. Sort Data")
        print("6. Display Dataset Statistics")
        print("7. Exit program ")

        options = input("Enter your option:")

        if options == '1':
            data = input_data()
        elif options == '2':
            data_summary(data)
        elif options == '3':
            calculate_factorial()
        elif options == '4':
            filter_data(data)
        elif options == '5':
            sort_data(data)
        elif options == '6':
            dataset_statistics(data)
        elif options == '7':
            print("Thank you for using the Data Analyzer and Transformer Program. Goodbye!")
            break
        else:
            print("Invalid Opitions,Please enter valid options")
menuu()
```
