```python
import datetime
import time
import math
import random
import uuid

# DATETIME OPERATIONS
# ---------------------------
def datetime_menu():
    while True:
        print("""
datetime and time operations:
1. Display current date and time
2. Calculate difference between two dates/times
3. Format date into custom format
4. Stopwatch
5. Countdown timer
6. Back to main menu
""")
        choice = input("Enter your choice : ")

        # 1. current datetime
        if choice == "1":
            now = datetime.datetime.now()
            print(f"\ncurrent date and time : {now.strftime('%Y-%m-%d %H:%M:%S')}")
            print("=======================")

        # 2. difference between dates
        elif choice == "2":
            d1 = input("Enter the first date (YYYY-MM-DD) : ")
            d2 = input("Enter the second date (YYYY-MM-DD) : ")

            d1 = datetime.datetime.strptime(d1, "%Y-%m-%d")
            d2 = datetime.datetime.strptime(d2, "%Y-%m-%d")
            diff = abs((d2 - d1).days)

            print(f"Difference : {diff} days")

        # 3. format date
        elif choice == "3":
            date_in = input("Enter date (YYYY-MM-DD): ")
            fmt = input("Enter desired format (e.g. %d/%m/%Y): ")
            d = datetime.datetime.strptime(date_in, "%Y-%m-%d")
            print("Formatted date:", d.strftime(fmt))

        # 4. stopwatch
        elif choice == "4":
            input("Press Enter to start stopwatch...")
            start = time.time()
            input("Press Enter to stop stopwatch...")
            end = time.time()
            print(f"Elapsed time: {round(end - start, 2)} seconds")

        # 5. countdown timer
        elif choice == "5":
            sec = int(input("Enter countdown time (seconds): "))
            while sec > 0:
                print(sec, end="\r")
                time.sleep(1)
                sec -= 1
            print("Time's up!")

        elif choice == "6":
            return
        
        else:
            print("Invalid choice!")

# MATHEMATICAL OPERATIONS
# ---------------------------
def math_menu():
    while True:
        print("""
Mathematical operations :
1. Calculate factorial
2. Solve compound interest
3. Trigonometric calculations
4. Area of geometric shapes
5. Back to main menu 
""")
        choice = input("Enter your choice : ")

        # 1. factorial
        if choice == "1":
            n = int(input("Enter a number : "))
            print("Factorial :", math.factorial(n))

        # 2. compound interest
        elif choice == "2":
            p = float(input("Enter principal amount : "))
            r = float(input("Enter rate of interest (in %) : "))
            t = float(input("Enter time (in years) : "))
            ci = p * (1 + r/100)**t
            print("Compound Interest : %.2f" % ci)

        # 3. trigonometric
        elif choice == "3":
            angle = float(input("Enter angle in degrees: "))
            rad = math.radians(angle)
            print("sin:", math.sin(rad))
            print("cos:", math.cos(rad))
            print("tan:", math.tan(rad))

        # 4. area calculation
        elif choice == "4":
            print("""
Shapes:
1. Circle
2. Rectangle
3. Triangle
""")
            s = input("Choose shape: ")
            if s == "1":
                r = float(input("Enter radius: "))
                print("Area:", math.pi * r * r)
            elif s == "2":
                a = float(input("Enter length: "))
                b = float(input("Enter breadth: "))
                print("Area:", a * b)
            elif s == "3":
                b = float(input("Enter base: "))
                h = float(input("Enter height: "))
                print("Area:", 0.5 * b * h)

        elif choice == "5":
            return
        
        else:
            print("Invalid choice!")

# RANDOM DATA GENERATION
# ---------------------------
def random_menu():
    while True:
        print("""
Random Data Generation :
1. Generate Random Number
2. Generate Random List
3. Create Random Password
4. Generate Random OTP
5. Back to Main Menu
""")
        choice = input("Enter your choice : ")

        # 1. random number
        if choice == "1":
            print("Random number :", random.randint(1, 100))

        # 2. random list
        elif choice == "2":
            l = [random.randint(1, 100) for _ in range(10)]
            print("Random List :", l)

        # 3. random password
        elif choice == "3":
            length = int(input("Enter password length : "))
            chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789@#$%&"
            pwd = ''.join(random.choice(chars) for _ in range(length))
            print("generated password :", pwd)

        # 4. OTP
        elif choice == "4":
            print("Generated OTP :", random.randint(100000, 999999))

        elif choice == "5":
            return
        
        else:
            print("Invalid choice!")

# UUID GENERATOR
# ---------------------------
def uuid_menu():
    print("\nGenerated UUID :", uuid.uuid4())

# FILE OPERATIONS
# ---------------------------
def file_menu():
    while True:
        print("""
File operations:
1. Create a new file
2. Write to a file
3. Read from a file
4. Append to a file
5. Back to Main Menu 
""")
        choice = input("Enter your choice : ")

        # 1. create
        if choice == "1":
            name = input("Enter file name : ")
            open(name, "w").close()
            print("File created successfully!")

        # 2. write
        elif choice == "2":
            name = input("Enter file name : ")
            data = input("Enter data to write : ")
            with open(name, "w") as f:
                f.write(data)
            print("Data written successfully !")

        # 3. read
        elif choice == "3":
            name = input("Enter file name : ")
            try:
                with open(name, "r") as f:
                    print("File Content :")
                    print(f.read())
            except:
                print("File not found.")

        # 4. append
        elif choice == "4":
            name = input("Enter file name : ")
            data = input("Enter data to append : ")
            with open(name, "a") as f:
                f.write(data)
            print("Data appended successfully !")

        elif choice == "5":
            return

        else:
            print("Invalid choice!")

# EXPLORE MODULE ATTRIBUTES
# ---------------------------
def module_menu():
    name = input("Enter module name to explore : ")
    try:
        mod = __import__(name)
        print("Available Attributes in", name, "module :")
        print(dir(mod))
    except:
        print("Invalid module name!")

# MAIN MENU
# ---------------------------
while True:
    print("""
choose an option:
1. Datetime and Time operations
2. Mathematical operations
3. Random data generation 
4. Generate unique identifiers(UUID)
5. File operations
6. Explore module attributes(dir())
7. Exit
======================
""")

    main_choice = input("Enter your choice : ")

    if main_choice == "1":
        datetime_menu()
    elif main_choice == "2":
        math_menu()
    elif main_choice == "3":
        random_menu()
    elif main_choice == "4":
        uuid_menu()
    elif main_choice == "5":
        file_menu()
    elif main_choice == "6":
        module_menu()
    elif main_choice == "7":
        print("Thank you for using the multi-utility toolkit!")
        print("============================")
        break
    else:
        print("Invalid choice!")
```
